```python
students_indices = data[data['role'] == 'student'].index
alumni_indices = data[data['role'] == 'alumni'].index
```

```python
student_vectors = vectors[students_indices]
alumni_vectors = vectors[alumni_indices]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarities = cosine_similarity(student_vectors, alumni_vectors)
```

```python
def get_recommendations(student_index):
    student = data.loc[student_index, 'tags']
    distances = similarities[student_index]
    #alumni_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    top_alumni_indices = sorted(zip(alumni_indices, distances), key=lambda x: x[1], reverse=True)[:5]
    #print("Student: ", data.iloc[student_index].skills)

    print("\n🎯 **Student Skills:**", ', '.join(data.loc[student_index, 'skills']))

    print("\n🔍 **Top 5 Matching Alumni:**")
    for alum_index, score in top_alumni_indices:
        print(f"⭐ {', '.join(data.loc[alum_index, 'skills'])} (Similarity: {score:.3f})")
```

```python
get_recommendations(2)
```
